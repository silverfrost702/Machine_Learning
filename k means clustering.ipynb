```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,random_state=101)
```

```python
#data[0]
plt.xlim(-20,10)
plt.ylim(-15,15)
plt.scatter(data[0][:,0], data[0][:,1])
```

```python
data[1] # these are the centres
```

```python
plt.xlim(-20,10)
plt.ylim(-15,15)
plt.scatter(data[0][:,0], data[0][:,1], c = data[1], cmap = 'plasma')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])           # fit to the features
```

```python
kmeans.cluster_centers_
```

```python
 kmeans.labels_      #these are the predicted labels
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, sharey = True, figsize = (10,6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap = 'plasma')

ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap = 'plasma')
```
